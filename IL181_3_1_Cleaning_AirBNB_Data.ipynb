# Importing the Data

```
import pandas as pd

data = pd.read_csv("/content/drive/My Drive/Minerva Year 4 - Synthesis - London SF/IL181 - Practical Data Science/AB_NYC_2019.csv")
```

# Cleaning The Data

```
data = data.drop(['name', 'id'], axis=1)
```

```
neighbourhood = pd.get_dummies(data['neighbourhood'])
neighbourhood_group = pd.get_dummies(data['neighbourhood_group'])
room = pd.get_dummies(data['room_type'])
data = pd.concat([data, neighbourhood], axis=1)
data = pd.concat([data, neighbourhood_group], axis=1)
data = pd.concat([data, room], axis=1)
data = data.drop(["neighbourhood_group", "neighbourhood", "room_type"], axis=1)
```

```
data = data.dropna(axis=0)
data['last_review'] = pd.to_datetime(data['last_review'])
```

```
y = pd.DataFrame(data['price'])
X = pd.DataFrame(data.drop(['host_id', 'host_name', 'price', 'last_review'], axis=1))
```

```
from sklearn import svm
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```
